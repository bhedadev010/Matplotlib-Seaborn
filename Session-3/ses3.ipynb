```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-virginica':0,'Iris-setosa':1,'Iris-versicolor':2})
iris.sample(5)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['SepalWidthCm'],c=iris['Species'],cmap='jet',alpha=0.3)
plt.colorbar()
```

```python
df= pd.read_csv('batter.csv')
df = df.head(50)
```

```python
plt.scatter(df['avg'],df['runs'],s=df['runs']*0.3,alpha=0.3)

for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['runs'].values[i],df['batter'].values[i])


plt.title('avg vs run')
plt.xlabel('avg')
plt.ylabel('run')
plt.figure(figsize=(15,6))
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(12,6))

ax[0].scatter(df['avg'],df['strike_rate'])
ax[1].scatter(df['avg'],df['runs'])

ax[0].set_title('avg vs sr')
ax[1].set_title('avg vs runs')
```

```python
fig,ax = plt.subplots(2,2,sharex=True,figsize=(12,6))

ax[0,0].scatter(df['avg'],df['strike_rate'])
ax[0,1].scatter(df['avg'],df['runs'])
ax[1,0].hist(df['runs'])
ax[1,1].hist(df['avg'])
```

```python
fig = plt.figure(figsize=(10,6))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(df['avg'],df['strike_rate'])

ax1 = fig.add_subplot(2,2,2)
ax1.hist(df['strike_rate'])

ax1 = fig.add_subplot(2,2,3)
ax1.hist(df['avg'])
```
