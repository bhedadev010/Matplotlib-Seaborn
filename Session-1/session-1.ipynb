```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dotted',marker='D')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dotted',marker='o')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dotted',marker='o')
plt.title('sharma vs kholi')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linewidth=3,linestyle='dotted',marker='o',label='v kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dotted',marker='o',label='sharma')
plt.title('sharma vs kholi')
plt.xlabel('season')
plt.ylabel('runs')

plt.legend(loc='upper left')
plt.grid()
```
