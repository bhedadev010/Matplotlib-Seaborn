```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# axes level -> scatter plot
sns.scatterplot(data = tips, x='total_bill',y='tip',hue='sex',size='size',style='time')
```

```python
# figure level-> relational plot(relplot)
sns.relplot(data=tips, x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```
