```python
plt.plot(batsman['index'],batsman['V Kohli'],marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('RG Sharma VS Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dotted',marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed',marker='o')
plt.title('RG Sharma VS Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
# linestyle =  'solid', 'dashed', 'dashdot', 'dotted'
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linewidth=5,marker='D',label='Virat Kohli',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed',marker='o',label='Rohit Sharma')
plt.title('RG Sharma VS Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(loc='lower right')
plt.grid()
```

```python
df=pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='green')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015']+df['2016']),label='2017')

plt.legend()
```
