```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
tips
```

```python
sns.scatterplot(data=tips,x='day',y='total_bill')
```

```python
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
sns.catplot(data=tips,x='total_bill',kind='box')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
```

```python
sns.barplot(data=tips,x='sex',y='total_bill')
plt.grid()
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker')
plt.grid()
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)
plt.grid()
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',ci=None)
plt.grid()
```

```python
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker')
```
