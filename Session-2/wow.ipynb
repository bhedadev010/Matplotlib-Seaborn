```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
axs[0, 0].plot([1, 2, 3, 4], [1, 4, 2, 3], 'r')
axs[0, 1].scatter([1, 2, 3, 4], [4, 3, 2, 1], color='b')
axs[1, 0].bar([1, 2, 3, 4], [1, 4, 9, 16])
axs[1, 1].hist([1, 2, 1, 3, 2, 4, 3, 2], bins=4)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)
ax.plot(x, y, z)
plt.show()
```

```python
import numpy as np

data = np.random.rand(10, 10)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
import matplotlib.colors as mcolors

colors = ["#d73027", "#f46d43", "#fdae61", "#fee08b", "#d9ef8b", "#a6d96a", "#66bd63", "#1a9850"]
custom_cmap = mcolors.LinearSegmentedColormap.from_list("custom", colors)
plt.imshow(data, cmap=custom_cmap)
plt.colorbar()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.annotate('local max', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 0.8),
             arrowprops=dict())
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(['One', 'Two', 'Three', 'Four'], rotation=45, fontsize=12)
ax.set_yticks([10, 20, 25, 30])
plt.show()
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(2 * theta))

plt.polar(theta, r)
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 5)]
plt.violinplot(data)
plt.show()
```

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.streamplot(X, Y, U, V, color='r')
plt.show()
```
